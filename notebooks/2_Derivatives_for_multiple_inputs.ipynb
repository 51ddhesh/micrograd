```python
h = 0.0001
```

```python
da = ((a + h) * b) + c
print(f'The derivative of d wrt a is: {(da - d) / h}')
```

```python
db = (a * (b + h)) + c
print(f'The derivative of d wrt b is: {(db - d) / h}')
```

```python
dc = (a * b) + (c + h)
print(f'The derivative of d wrt c is: {(dc - d) / h}')
```
