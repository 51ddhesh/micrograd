```python
import numpy as np
import math
import matplotlib.pyplot as plt
from graphviz import Digraph
%matplotlib inline
```

```python
class Value:
    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self._prev = set(_children)
        self.grad = 0
        self._op = _op
        self.label = label

    def __repr__(self):
        return f'Value(data={self.data})'
    
    def __add__(self, other):
        return Value(self.data + other.data, (self, other), '+')
    
    def __mul__(self, other):
        return Value(self.data * other.data, (self, other), '*')
```

```python
def trace(root):
    nodes, edges = set(), set()
    def build(v):
        if v not in nodes:
            nodes.add(v)
            for child in v._prev:
                edges.add((child, v))
                build(child)
    build(root)
    return nodes, edges

def draw_dot(root):
    dot = Digraph(format='svg', graph_attr={'rankdir':'LR'})
    nodes, edges = trace(root)
    for n in nodes:
        uid = str(id(n))
        dot.node(name=uid, label='{ %s | data: %.4f | grad: %.4f }' % (n.label, n.data, n.grad), shape='record')
        if n._op:
            dot.node(name=uid + n._op, label=n._op)
            dot.edge(uid + n._op, uid)

    for n1, n2 in edges:
        dot.edge(str(id(n1)), str(id(n2)) + n2._op)

    return dot
```

```python
a = Value(2.0, label='a')
b = Value(-3.0, label='b')
c = a * b; c.label='c'
d = Value(10.0, label='d')
e = c + d; e.label='e'
f = Value(-2.0, label='f')
L = e * f; L.label='L' # L as in loss 

a.grad = 6.0
b.grad = -4.0
c.grad = -2.0
d.grad = -2.0
e.grad = -2.0
f.grad = 4.0
L.grad = 1.0
```

```python
draw_dot(L)
```

We will now nudge the values of the leaf nodes (which generally we have control over) in the direction of the gradient to make the final value 'more positive'

```python
h = 0.01
a.data += h * a.grad
b.data += h * b.grad
d.data += h * d.grad
f.data += h * f.grad
```

Re-running the forward pass

```python
c = a * b
e = d + c
L = e * f
print(L)
```
